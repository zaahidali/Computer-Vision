## # Detect any simple BGR Color in an image using Color thresholding technique

```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input3.jpg',1)

d = img.shape
shape = (d[1]+200,d[0]+200)
img = cv2.resize(img,shape)

region = cv2.selectROI('window',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# colors in BGR form
(x,y,w,h) = region
cropped = img[y:y+h, x:x+w]
print(cropped)



```

```python



import numpy as np
import matplotlib.pyplot as plt

#bgr
blue = img[:,:,0]
green = img[:,:,1]
red = img[:,:,2]

#image rows & cols & 8 bit image
d = img.shape
m = d[0]
n = d[1]
o = np.zeros([m,n], dtype=np.uint8)

#for blueish color
for i in range(1,m):
    for j in range(1,n):
        if blue[i][j]==249 and green[i][j]==33 and red[i][j]== 4:
            o[i][j] = 255
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.axis("off")

plt.subplot(1,2,2) 
plt.imshow(o,'gray')
plt.title("Color Detected")
plt.axis("off")

plt.savefig("newImage.jpg")
plt.show()
```

# Edge Detection

```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('colors.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)

```

```python
img_canny = cv2.Canny(img,-100,150)

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.axis("off")

plt.subplot(1,2,2) 
plt.imshow(img_canny)
plt.title("Canny")
plt.axis("off")
plt.show()
```

```python
sobelX = cv2.Sobel(img,cv2.CV_8U,1,0)
sobelY = cv2.Sobel(img,-1,0,1)

im_sobel = sobelX + sobelY 
plt.imshow(im_sobel)

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.axis("off")

plt.subplot(1,2,2) 
plt.imshow(im_sobel)
plt.title("Sobel")
plt.axis("off")
plt.show()
```

```python

########### robert mask
mask1 = np.array([[1,0],[0,-1]])
mask2 = np.array([[0,1],[-1,0]])

img_rX = cv2.filter2D(img,cv2.CV_8U,mask1)
img_rY = cv2.filter2D(img,-1,mask2)
img_rob = img_rX +img_rY 
#plt.imshow(img_rob)


plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.axis("off")

plt.subplot(1,2,2) 
plt.imshow(img_rob)
plt.title("Robert")
plt.axis("off")

plt.show()
```

```python
mask1 = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
mask2 = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
prewX = cv2.filter2D(img,cv2.CV_8U,mask1)
prewY = cv2.filter2D(img,-1,mask2)
img_prew = prewX+ prewY

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.axis("off")

plt.subplot(1,2,2) 
plt.imshow(img_prew)
plt.title("Prewitts")
plt.axis("off")
plt.show()
```
